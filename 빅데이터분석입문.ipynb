```python
reports = {"tom":{"kor":60, "math": 100},
           "suzy":{"kor":70, "math": 90},
           "lina":{"kor":80, "math": 75},
           "alice":{"kor":90, "math": 85},
           "bomin":{"kor":100, "math":100},
           }
math_tot = 0
kor_tot = 0
class_num = 0
for student in reports.keys():
    class_num += 1
    kor_tot += reports[student]["kor"]
    math_tot += reports[student]["math"]

kor_mean = round(kor_tot/class_num, ndigits=1)
math_mean = round(math_tot/class_num, ndigits=1)

print(f'국어 평균은 {kor_mean}이고, 수학 평균은 {math_mean}입니다.')
```

```python
reports = {"tom":{"kor":60, "math": 100},
           "suzy":{"kor":70, "math": 90},
           "lina":{"kor":80, "math": 75},
           "alice":{"kor":90, "math": 85},
           "bomin":{"kor":100, "math":100},
           }
from numpy import mean
math_points = [] 
kor_points = []
for student in reports.keys():
        for subject in reports[student].keys():
            if subject == "kor":
                kor_points.append(reports[student]["kor"])
            elif subject == "math":
                math_points.append(reports[student]["math"])

print(f'국어 평균은 {mean(kor_points)}이고, 수학 평균은 {mean(math_points)}입니다.')
```

### class
* class를 사용하는 가장 중요한 이유는 재사용이 가능하며, 유지보수에 유리하다.
* . 연산자를 이용하는 것은 클래스에 접근하기 위하거나, 터널(메서드)를 이용하기 위해서이다.

```python
class Student:

    def __init__(self, name, korean, english, math):
        self.name = name
        self.korean = korean
        self.english = english
        self.math = math

    def get_sum(self):
        return self.korean + self.english + self.math

    def get_avg(self):
        return (self.get_sum() / 3)
    
    def __str__(self):
        return f' name : {self.name}'

ajh = Student('jaehyun', 100, 100 , 90)

print(ajh.get_sum())
print(ajh.get_avg())

print(ajh)


```
